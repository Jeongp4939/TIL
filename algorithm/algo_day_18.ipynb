```python
# 정사각형방

def bfs(si, sj):
    q = []  # [0] 생성
    alst = []
 
    q.append((si,sj))   # [1] 초기데이터 삽입
    v[si][sj]=1
    alst.append(arr[si][sj])
 
    while q:
        ci,cj = q.pop(0)
 
        # 4방향, 미방문, 조건 맞으면!! (1차이)
        for di,dj in ((-1,0),(1,0),(0,-1),(0,1)):
            ni,nj = ci+di, cj+dj
            if 0<=ni<N and 0<=nj<N and v[ni][nj]==0 and\
                abs(arr[ci][cj]-arr[ni][nj])==1:
                q.append((ni,nj))
                v[ni][nj]=1
                alst.append(arr[ni][nj])
 
    return min(alst), len(alst)
 
T = int(input())
for test_case in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]
    v = [[0]*N for _ in range(N)]
    ans, cnt = N*N, 0
    for si in range(N):
        for sj in range(N):
            if v[si][sj]==0:
                t, tcnt = bfs(si, sj)
                if cnt<tcnt or cnt==tcnt and ans>t:
                    ans, cnt = t, tcnt
 
    print(f'#{test_case} {ans} {cnt}')
```

# 홈 방법 서비스

모든 시작점, k를 넓혀가면서

for i (si-k+1,si+k)

    for j (sj-k+1+abs(si-i), sj+k-abs(si-i))
    
범위 이내, 집이면 cnt+=1

```python
# for문(loop)으로 풀이
cost=[(k**2+(k+1)**2) for k in range(40)] 
def solve_loop():
    for si in range(N):
        for sj in range(N): # 가능한 모든 시작 위치
            for k in range(1,2*N):
                cnt = 0
                for i in range(si-k+1, si+k):
                    t = abs(si-i)
                    for j in range(sj-k+1+t, sj+k-t):
                        if 0<=i<N and 0<=j<N:
                            cnt+= arr[i][j] # 집위치를 더하기 (집이 1이니까)
                # 운영비용 보다 수익이 같거나 큰 경우 정답 갱신
                if cost[k]<cnt*M:
                    mx = max(mx,cnt)
    return mx

T = int(input())
for tc in range(1,1+T):
    N,M = map(int,input().split())  # M:비용
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = solve_loop()
    print(f'#{tc} {ans}')
```

```python
# bfs 풀이 코드

def bfs(si,sj):
    q=[]
    v=[[0]*N for _ in range(N)]
    old = 0


    q.append((si,sj))
    v[si][sj] = 1
    cnt = arr[si][sj]   # 시작좌표가 집이면 1, 아니면 0

    while q:
        ci,cj = q.pop(0)
        if old != v[ci][cj]:    # k값이 달라진 경우
            old = v[ci][cj]
            if cost[v[ci][cj]]<=cnt*M:
                mx = max(mx,cnt)


        for di,dj in ((-1,0),(0,-1),(1,0),(0,1)):
            ni,nj = ci+di, cj+dj
            if 0<=ni<N and 0<=nj<N and v[ni][nj]==0:
                q.append((ni,nj))
                v[ni][nj]=v[ci][cj]+1
                cnt+=arr[ni][nj]


def solve_bfs():
    mx = 0
    for si in range(N):
        for sj in range(N): # 가능한 모든 시작 위치
            mx = max(mx,bfs(si,sj))
    return mx


T = int(input())
for tc in range(1,1+T):
    N,M = map(int,input().split())  # M:비용
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = solve_loop()
    print(f'#{tc} {ans}')
```
