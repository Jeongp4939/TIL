- 경계 지점에서의 조건을 매번 확인 해야할까?
    - 필요에 따라 array 바깥을 특정 값으로 처리하여 범위 확인 등 연산을 줄일 수 있음

````python
def count(arr):
    ans = 0
    for lst in arr:
        cnt=0
        for n in lst:
            if n ==1:       # 단어를 넣을 수 있는 공백
                cnt+=1
            else:           # 칸 없음!
                if cnt == K: 
                    ans+=1
                cnt=0
    return ans

T = int(input())
for tc in range(1,T+1):
    N, K = map(int,input().split())
    arr = [list(map(int, input().split())) + [0] + [0] for _ in range(N) + [[0]*(N+1)]]
    # arr과 arr_t로 각각 개수를 계산
    arr_t = list(map(list,zip(*arr)))
    ans = count(arr) + count(arr_t)
    print(f'#{tc} {ans}')
````

##### 스도쿠 검증
- 한줄, 모든 숫자가 한번만 나와야 함
    ```python
    lst = [7,3,6,4,2,1,5,8,9]
    len(set(lst))!=9    # 스도쿠x
    # 열-> 행으로 바꿔서 검사
    # 사각형 검사
    for i in [0,3,6]:
        for i in [0,3,6]:   # 가능한 모든 시작
            lst = arr[i][j:j+3] + arr[i+1][j:j+3] + arr[i+2][j:j+3]
    len(set(lst))!=9: # X
    
    ```

````python
def solve(arr):
    for lst in arr:             # 행을 체크
        if len(set(lst))!=9:    # 스도쿠 실패
            return 0
    
    arr_t = list(zip(*arr))
    for lst in arr_t:           # 열을 체크
        if len(set(lst))!=9:    # 스도쿠 실패
            return 0
    
    for i in (0,3,6):
        for j in (0,3,6):       # 3*3 격자
            lst = arr[i][j:j+3] + arr[i+1][j:j+3] + arr[i+2][j:j+3]
            if len(set(lst))!=9:
                return 0
    return 1

T = int(input())
for tc in range(1,T+1):
    arr = [list(map(int, input().split())) for _ in range(9)]
    ans = solve(arr)
    print(f'#{tc} {ans}')
````

### 숫자 배열 회전
- 90도 회전 규칙성의 발견
    - 목적지 기준으로 원본의 어떤 좌표의 값이 저장되는지 확인
    - 목적지 기준 i,j 범위를 기준으로 정리

    ```python
    arr1[i][j] = arr[N-1-j][i]
    ```

- 180도 회전
    ```python
    arr2[i][j] = arr[N-1-i][N-1-j]
    ```

- 270도 회전
    ```python
    arr3[i][j] = arr[j][N-1-i]
    ```

```python
for a,b,c in zip(a1,a2,a3):
    {"".join(a)} {"".join(b)} 
```

````python
T = int(input())
for tc in range(1,T+1):
    N = int(input())
    arr = [input().split() for _ in range(N)]
    a1 = [[0]*N for _ in range(N)]  # 90도
    a2 = [[0]*N for _ in range(N)]
    a3 = [[0]*N for _ in range(N)]
    
    # 회전각도에 따른 위치값을 저장
    for i in range(N):
        for j in range(N):
            a1[i][j] = arr[N-1-j][i]
            a2[i][j] = arr[N-1-i][N-1-j]
            a3[i][j] = arr[j][N-1-i]
    
    print(f'#{tc}')
    
    for a,b,c in zip(a1,a2,a3):
        print(f'{"".join(a)} {"".join(b)} {"".join(c)}')
````

# ladder2
- 좌,우 1순위, 진행방향(dj(-1, 1, 0->아래로[초기값])) 직진
    - 1~100까지(양 끝에 패딩을 추가)
    - si<-0, sj(1인 지점), cnt, ci, cj
        ```python
        while ci<99:    # 99가 되면 종료
            if dj==0:   # 아래방향(좌/우)
                if arr[ci][cj-1]==1: #좌
                    dj=-1, cj-=1
                elif arr[ci][cj+1]==1: # 우
                    dj=1, cj+=1
                else:                   # 아래
                    ci+=1
            else:   # 좌/우 -> 직진(진행방향 1이면)
                if arr[cj][cj+dj]==1: cj+=dj
                else: dj=0, ci+=1 # 아래
        if mx>=cnt: # 정답 갱신 sj 좌표
            mx,ans = cnt, sj-1
        ```

- 범위 내 체크
    - sj-1
    - ans-1
    



````python
T = 10
for test_case in range(1, T + 1):
    _ = int(input())
    arr = [[0] + list(map(int, input().split())) + [0] for _ in range(100)]
    mn = 100*100
    for sj in range(1, 101):
        # [1] 시작지점 찾기
        si = 0
        if arr[si][sj]!=1:
            continue
        cnt, dj = 0, 0
        ci, cj = si, sj
        while ci<99:
            cnt+=1
            if dj==0:
                if arr[ci][cj-1]==1:    #좌측
                    dj=-1
                    cj-=1
                elif arr[ci][cj+1]==1:  #우측
                    dj=1
                    cj+=1
                else:
                    ci+=1
            else:
                if arr[ci][cj+dj]==1:
                    cj+=dj
                else:   # 진행방향이 막힌경우 아래로
                    dj=0
                    ci+=1
        if mn>=cnt:
            mn, ans = cnt, sj-1
 
    print(f'#{test_case} {ans}')
````

````python
# 누적합 출력하기
# 출력 결과 : 3 7 12 13 19 28 19 13 12 7 3

# 1. sum을 매개 변수로 놓고 코드 작성

arr = [3,4,5,1,6,9]

def a_sum(level, sum):
    print(sum,end=' ')
    if level == len(arr)-1:
        return
    a_sum(level+1,sum+arr[level+1])
    print(sum,end=' ')
    
a_sum(0,arr[0])
    
print()
# 2. sum을 전역 변수로 놓고 코드 작성

sum = arr[0]

def a_sum_global(level):
    global sum
    print(sum,end=' ')
    if level==5:
        return
    sum+=arr[level+1]
    a_sum_global(level+1)
    sum-=arr[level+1]
    print(sum,end=' ')

a_sum_global(0)
````

````python
# pond
# 0번 인덱스에 개구리(idx에 해당하는 숫자 칸씩 점프)

arr = [2,0,1,1,3,5,1]

def jump(idx):
    if idx >= len(arr):         # 벗어나면 종료
        return
    jump(idx+arr[idx])
    print(arr[idx], end=' ')    # 뒤에서 부터 출력
    
    
jump(0)
````

````python
card = [3,7,1,2]
sum = 0

def abc(level):
    global sum
    if level==3:
        print(sum, end=' ')
        return
    for i in range(4):
        sum+=card[i]
        abc(level+1)
        sum-=card[i]

abc(0)
````

````python
# arr = [2,7,1,4,3]
# n개의 카드 묶음이 존재(arr로 이루어징)
# n개의 카드 묶음에서 각각 1장씩 뽑아서 더했을 때 합이 20이 넘는 경우는 몇가지?
# level은 n까지
# branch는 2,7,1,4,3 5개
# sum 매개변수로 선언
# cnt 전역 변수로 설정
# sum>20 일 경우 cnt ++

n = int(input())
arr = [2,7,1,4,3]
cnt = 0

def abc(level, sum=0):
    global cnt
    
    if level == n:
        if sum>20:
            cnt+=1
        return
    for i in range(5):
        abc(level+1, sum+arr[i])
        
abc(0)
print(cnt)
    
````

````python
# 거슬러 줄 돈 changes 입력
# 사용할 수 있는 동전 110,70,10원

coins = [110,70,10]

changes = int(input())
m_coin = 10**10

def abc(level,changes):
    global m_coin
    if changes < 0 :
        return
    if changes == 0:
        if level<m_coin:
            m_coin=level
            return
    for i in range(3):
        abc(level+1, changes-coins[i])

abc(0,changes)
print(m_coin)
        
    
    
````

````python
# A,B,C 두 뭉치
# 카드 한장 씩 뽑았을 때 나올 수 있는 모든 경우의 수
# path = [0]*4
# lvl = 2, br = 3
# path 전역
# if level == 2: return
# 들어갈 때 미리 path에 저장 후 진행

# n = int(input())
n = 2
path = ['']*n
card = ['A','B','C']

def abc(level):
    if level == n:
        print(''.join(path))
        return
    for i in range(3):
        path[level] = card[i]
        abc(level+1)
        #path[level]=''

abc(0)
````

````python
# n개의 주사위를 던졌을 때 나올 수 있는 모든 배열을 출력
# path = [0]*n
# dice = [1,2,3,4,5,6]
# level -> 주사위 던지는 횟수 -> n번에서 return

n = int(input())
path = [0]*n


def abc(level):
    if level == n:
        for i in range(n):
            print(path[i],end = ' ')
        print()
        return
    for i in range(1,7):    # 주사위 눈 1~6
        path[level] = i
        abc(level+1)

abc(0)
````
